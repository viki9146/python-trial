# visualizing data using graphs

```python
sns.set_style("whitegrid")
sns.pairplot(hue='Species', data=iris,height=3)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.title('Scatterplot of Iris Data',size = 25)
plt.scatter(iris['SepalLengthCm'],iris['SepalWidthCm'],iris['PetalLengthCm'],iris['PetalWidthCm'])
plt.show()
```

```python
x = iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
```

# using elbow method

```python
ss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(x)
    ss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.title('The Elbow Method')
sns.lineplot(range(1, 11),ss,marker='o',color='red')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=i,init='k-means++')
ymeans = kmeans.fit_predict(x) 
```

# visualising the model 

```python
plt.figure(figsize=(15,7))
sns.scatterplot(x[ymeans == 0, 0], x[ymeans == 0, 1], color = 'red', label = 'Cluster 1',s=50)
sns.scatterplot(x[ymeans == 1, 0], x[ymeans == 1, 1], color = 'blue', label = 'Cluster 2',s=50)
sns.scatterplot(x[ymeans == 2, 0], x[ymeans == 2, 1], color = 'green', label = 'Cluster 3',s=50)
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'black', 
                label = 'Centroids',s=300,marker=',')
```

# Evaluting the result

```python
#k-means
kmeans= KMeans(n_clusters=3)
kmeans_predict = kmeans.fit_predict(x)
## cross tabulation table for k-means
DataFrame1= pd.DataFrame({"labels":kmeans_predict,"Species":iris['Species']})
crt1=pd.crosstab(DataFrame1['labels'],DataFrame1['Species'])
```

```python
crt1
```

# visualize the result

```python
plt.figure(figsize=(30,10))
plt.suptitle("Cross Tabulation",fontsize=20)
plt.subplot(1,2,1)
plt.title('K-Means',fontsize=18)
sns.heatmap(crt1,annot=True,cbar=True,cmap='RdPu_r')
plt.show()
```
